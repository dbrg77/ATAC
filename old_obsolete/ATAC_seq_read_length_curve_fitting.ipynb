```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pysam
import matplotlib.mlab as mlab
from scipy.optimize import curve_fit
```

# Load data, and filter out unwanted fragments (either too large or too small).

```python
bam = pysam.Samfile("Th2_2h_f2_no_chrM_picard_markdup_cosorted.bam", 'rb')

frag_size = []

for read in bam:
    if bam.getrname(read.tid)!="chrM" and read.tlen>50 and read.tlen<1500:
        frag_size.append(read.tlen)
bam.close()

frag_size = np.array(frag_size)
```

# Have a look at the raw histogram of fragment length distribution

```python
num_bins = np.linspace(0,1500,1501)
plt.hist(frag_size, num_bins, histtype="step", normed=1)

plt.xticks(range(0,1600,200))
plt.xlim(0,1600)
plt.xlabel('Fragment length (bp)')
plt.ylabel('Density')
plt.subplots_adjust(left=0.15)
plt.show()
```

# Plot the histogram in log scale

```python
plt.hist(frag_size, num_bins, histtype="step", normed=1)

plt.xticks(range(0,1600,200))
plt.xlim(0,1600)
plt.yscale("log")
plt.xlabel('Fragment length (bp)')
plt.ylabel('Density')
plt.subplots_adjust(left=0.15)
plt.show()
```

# Have a look at the fragment length distribution in scatter plot

```python
y, scatter_x = np.histogram(frag_size, num_bins, normed=1)

# get the mid-point of each bin
x = (scatter_x[:-1]+scatter_x[1:])/2

plt.plot(x, y, '.')
plt.xticks(range(0,1600,200))
plt.xlim(0,1600)
plt.xlabel('Fragment length (bp)')
plt.ylabel('Density')
plt.subplots_adjust(left=0.15)
plt.show()
```

# Nucleosome associated reads fragment length curve fitting starts here

### First, define a function that return the sum of six different Gaussian distribution

```python
# The raw data looks like it contains 6 normal distributions

def gauss(x, *p):
    m1, s1, w1, m2, s2, w2, m3, s3, w3, m4, s4, w4, m5, s5, w5, m6, s6, w6 = p
    return (mlab.normpdf(x, m1, s1) * w1 +
            mlab.normpdf(x, m2, s2) * w2 +
            mlab.normpdf(x, m3, s3) * w3 +
            mlab.normpdf(x, m4, s4) * w4 +
            mlab.normpdf(x, m5, s5) * w5 +
            mlab.normpdf(x, m6, s6) * w6)
```

### Then start the fit with some guess by eyes

```python
# initial guess
p0 = [200, 50, 0.7, 400, 50, 0.15, 600, 50, 0.1, 800, 55, 0.045, 1000, 60, 0.015, 1200, 60, 0.007]

popt, pcov = curve_fit(gauss, x[151:], y[151:], p0=p0)

print popt
```

### Now plot the fitted data and original data

```python
m1, s1, w1, m2, s2, w2, m3, s3, w3, m4, s4, w4, m5, s5, w5, m6, s6, w6 = popt

plt.figure(figsize=(10,20))

# plot the original data and the fitted six different Gaussian distributions
plt.subplot(211)
plt.hist(frag_size, num_bins, histtype="step", ec="k", normed=1, alpha=0.5)
plt.plot(x[151:], mlab.normpdf(x[151:], m1, s1)*w1, 'r-', lw=2)
plt.plot(x[151:], mlab.normpdf(x[151:], m2, s2)*w2, 'g-', lw=2)
plt.plot(x[151:], mlab.normpdf(x[151:], m3, s3)*w3, 'b-', lw=2)
plt.plot(x[151:], mlab.normpdf(x[151:], m4, s4)*w4, 'c-', lw=2)
plt.plot(x[151:], mlab.normpdf(x[151:], m5, s5)*w5, 'm-', lw=2)
plt.plot(x[151:], mlab.normpdf(x[151:], m6, s6)*w6, 'y-', lw=2)
plt.xticks(range(0,1600,200))
plt.xlim(0,1600)

# plot the fitted data
plt.subplot(212)
ys = gauss(x[151:], *popt)
plt.plot(x[151:], ys, 'k-')
plt.show()
```

### Now plot the original data, the fitted Gaussians, and the sum of the fit altogether

```python
plt.figure(figsize=(12,12))

plt.hist(frag_size, num_bins, histtype="step", ec="k", normed=1, alpha=0.8)
plt.plot(x, mlab.normpdf(x, m1, s1)*w1, 'r-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m2, s2)*w2, 'g-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m3, s3)*w3, 'b-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m4, s4)*w4, 'c-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m5, s5)*w5, 'm-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m6, s6)*w6, 'y-', lw=1.5)

ys = gauss(x, m1, s1, w1, m2, s2, w2, m3, s3, w3, m4, s4, w4, m5, s5, w5, m6, s6, w6)
plt.plot(x, ys, 'k--', lw=3)

plt.xticks(range(0,1600,200))
plt.xlim(0,1600)
#plt.ylim(0, 0.001)

plt.show()
```

```python
# define an exponetial function to fit the nucleosome-free reads

def expo(x, q, r):
    return q*np.exp(-r*x)
```

```python
popt2, pcov2 = curve_fit(expo, x[51:151], y[51:151], p0=[0.1, 0.05])

print popt2
```

### Now plot the fitted data and original data

```python
plt.plot(x[51:151], y[51:151], '.')
plt.plot(x[51:], expo(x[51:], *popt2), 'r-')
plt.xlim(0,1600)
plt.xlabel('Fragment length (bp)')
plt.ylabel('Density')
plt.subplots_adjust(left=0.15)
plt.show()
```

# Fit the whole data altogether

```python
# define a mixture function with 6 Gaussian distributions and one exponential distribution

def f(x, *p):
    m1, s1, w1, m2, s2, w2, m3, s3, w3, m4, s4, w4, m5, s5, w5, m6, s6, w6, q, r = p
    return (mlab.normpdf(x, m1, s1) * w1 +
            mlab.normpdf(x, m2, s2) * w2 +
            mlab.normpdf(x, m3, s3) * w3 +
            mlab.normpdf(x, m4, s4) * w4 +
            mlab.normpdf(x, m5, s5) * w5 +
            mlab.normpdf(x, m6, s6) * w6 +
            q*np.exp(-r*x))
```

```python
# rough initial guess based on previous results

guess = [200, 50, 0.7, 400, 50, 0.15, 600, 50, 0.1, 800, 55, 0.045, 1000, 60, 0.015, 1200, 60, 0.007, 0.03616429, 0.02297901]
```

```python
popt3, pcov3 = curve_fit(f, x[51:], y[51:], p0=guess, maxfev=100000)

print popt3
```

```python
m1, s1, w1, m2, s2, w2, m3, s3, w3, m4, s4, w4, m5, s5, w5, m6, s6, w6, q, r = popt3

plt.figure(figsize=(12,12))
plt.hist(frag_size, num_bins, histtype="step", ec="k", normed=1, alpha=0.5)
plt.plot(x, mlab.normpdf(x, m1, s1)*w1, 'r-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m2, s2)*w2, 'g-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m3, s3)*w3, 'b-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m4, s4)*w4, 'c-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m5, s5)*w5, 'm-', lw=1.5)
plt.plot(x, mlab.normpdf(x, m6, s6)*w6, 'y-', lw=1.5)
plt.plot(x, expo(x, 6.93385577e-02, 3.34278108e-02), 'k-', lw=1.5)

ys = f(x, *popt3)
plt.plot(x, ys, 'k--', lw=3.5)

plt.ylim(0, 0.015)
plt.show()
```
