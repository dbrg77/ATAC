```python
zRaw = np.loadtxt("/Users/xichen/Dropbox/Useful_small_files/Time_series_ATAC_seq/zNormalised_read_count_in_peaks_Matrix.txt",
                  skiprows=1, usecols=xrange(1,7))
tRaw = np.loadtxt("/Users/xichen/Dropbox/Useful_small_files/Time_series_ATAC_seq/tNormalised_read_count_in_peaks_Matrix.txt",
                  skiprows=1, usecols=xrange(1,7))
pid = np.loadtxt("/Users/xichen/Dropbox/Useful_small_files/Time_series_ATAC_seq/zNormalised_read_count_in_peaks_Matrix.txt",
                 skiprows=1, usecols=[0], dtype=np.str)
zData = zRaw.transpose()
tData = tRaw.transpose()
zScale = preprocessing.scale(zData, with_std=False)
tScale = preprocessing.scale(tData, with_std=False)
```

```python
pca1 = PCA(n_components=2)
pca1.fit(zScale)
zData_r = pca1.transform(zScale)

fig1 = plt.figure(figsize=(16,8))

ax1 = fig1.add_subplot(121)

ax1.scatter(zData_r[0:1, 0], zData_r[0:1, 1], marker='o', c='r', label="Naive", s=50, alpha=0.6)
ax1.scatter(zData_r[1:2, 0], zData_r[1:2, 1], marker='o', c='g', label="2h", s=50, alpha=0.6)
ax1.scatter(zData_r[2:3, 0], zData_r[2:3, 1], marker='o', c='b', label="4h", s=50, alpha=0.6)
ax1.scatter(zData_r[3:4, 0], zData_r[3:4, 1], marker='o', c='c', label="24h", s=50, alpha=0.6)
ax1.scatter(zData_r[4:5, 0], zData_r[4:5, 1], marker='o', c='m', label="48h", s=50, alpha=0.6)
ax1.scatter(zData_r[5:6, 0], zData_r[5:6, 1], marker='o', c='y', label="72h", s=50, alpha=0.6)

ax1.set_xlabel("PC1 (%.3f)" % pca1.explained_variance_ratio_[0])
ax1.set_ylabel("PC2 (%.3f)" % pca1.explained_variance_ratio_[1])
ax1.set_title("zNormalisation", fontsize=15)
ax1.legend(loc="upper right", ncol=3, scatterpoints=1, fontsize=10)

pca2 = PCA(n_components=2)
pca2.fit(tScale)
tData_r = pca2.transform(tScale)

ax2 = fig1.add_subplot(122)

ax2.scatter(tData_r[0:1, 0], tData_r[0:1, 1], marker='o', c='r', label="Naive", s=50, alpha=0.6)
ax2.scatter(tData_r[1:2, 0], tData_r[1:2, 1], marker='o', c='g', label="2h", s=50, alpha=0.6)
ax2.scatter(tData_r[2:3, 0], tData_r[2:3, 1], marker='o', c='b', label="4h", s=50, alpha=0.6)
ax2.scatter(tData_r[3:4, 0], tData_r[3:4, 1], marker='o', c='c', label="24h", s=50, alpha=0.6)
ax2.scatter(tData_r[4:5, 0], tData_r[4:5, 1], marker='o', c='m', label="48h", s=50, alpha=0.6)
ax2.scatter(tData_r[5:6, 0], tData_r[5:6, 1], marker='o', c='y', label="72h", s=50, alpha=0.6)

ax2.set_xlabel("PC1 (%.3f)" % pca2.explained_variance_ratio_[0])
ax2.set_ylabel("PC2 (%.3f)" % pca2.explained_variance_ratio_[1])
ax2.set_title("tNormalisation", fontsize=15)
ax2.legend(loc="upper right", ncol=3, scatterpoints=1, fontsize=10)

plt.show()
```
